```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from numpy.polynomial import Polynomial
```

```python
#2 Méthode d'interpolation de Lagrange
#Question 1

def base_lagrange(X, i):
    L = Polynomial([1])
    for j in range(len(X)):
        if j != i:
            a0 = -X[j, 0] / (X[i, 0] - X[j, 0])
            a1 = 1 / (X[i, 0] - X[j, 0])
            L = L * Polynomial([a0, a1])
    return L
```

```python
#Question 2 ( I DID IT MYSELF <3 )
def polynome_lagrange(X, Y):
    P = Polynomial([0])
    for i in range(len(X)):
        P += Y[i, 0]*base_lagrange(X, i)
    return P
```

```python
#Question 3
X = np.array([[-1, 0, 1]]).T #.T means transpose()
Y = np.array([[2, 1, -1]]).T
polynome_lagrange(X, Y)
np.arange(0, 5+1)
```

```python
#3 Approximation polynomiale au sens des moindres carrées
#Question 1

def matrice(X, p):
    A = np.zeros((len(X), p+1))
    for i in range(len(X)):
        for j in range(p+1):
            A[i,j] = X[i] ** j
    return A

X = [1, 4, 2]
matrice(X, 2)
```

```python
#Question 2
def regression(X, Y, p):
    A = matrice(X,p)
    return np.linalg.inv((A.T).dot(A)).dot((A.T).dot(Y))
```

```python
#Question 3
X = [-1, 0, 1]
Y = [2, 1, -1]
Polynomial(regression(X, Y, 1))
```
